```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
            ]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
             }
```

```python
def list_print():
    
    result = [];
    number_list = {}
    for i,item in enumerate(documents):
        directory_number = ''
        for shelf, docs in directories.items():
            if item['number'] in docs:
                directory_number = shelf
        number_list[item['number']] = {'type': item['type'],'name':item['name'],'directory': directory_number}

    for doc, info in number_list.items():
        result.append(f" №{doc}, тип:{info['type']}, владелец: {info['name']}, полка хранения {info['directory']}")

    return print((('\n').join(result)))
          

###--------------------------------------
def doc_checking (num):
    """
    Функция doc_checking позволяет проверить, есть ли вводимый документ в documents
    0 - значит документа нет
    1 - документ есть
    Требует переменную на вход в формате string
    """
    doc_check = 0
    for each in documents:
        if num == each['number']:
            doc_check = 1
    return doc_check
###--------------------------------------
def shelf_checking (num):
    """
    Функция shelf_checking позволяет проверить, существует указанная полка в directories
    0 - значит полки нет
    1 - полка есть
    Требует переменную на вход в формате string
    """
    result = 0
    if num in directories.keys():
        result = 1
    return result
###--------------------------------------
def move_doc():
    doc = str(input('введите номер документа '))
    if doc_checking(doc) == 0:
        print(f'Документ не найден в базе. Текущий список документов:')
        list_print()
    else:
        shelf = input('введите номер полки ')
        if shelf_checking(shelf) == 0:
            print (f"Такой полки не существует. Текущий перечень полок: {(','.join(list(directories.keys())))}")
        else:
            for shelf_dir, docs_dir in directories.items():
                if doc in shelf_dir:
                    doc_index = directories[shelf_dir].index(doc)
                    directories[shelf_dir].pop(doc_index)
                if shelf_dir == shelf:
                    directories[shelf].append(doc)
            print(f'Документ перемещен. Текущий список документов: ')
            list_print()
###--------------------------------------
def add_shelf():
    shelf = str(input('введите номер полки '))
    if shelf_checking(shelf) == 1:
        print (f"Такая полка существует. Текущий перечень полок: {(','.join(list(directories.keys())))}")
    else:
        directories.setdefault(shelf,[])
        print (f"Полка добавлена. Текущий перечень полок: {(','.join(list(directories.keys())))}")
###--------------------------------------    
def people ():
    doc = input('введите номер договора ')
    
    if doc_checking(doc) ==0:
        print(('Такого документа нет'))
    else:
        file={}
        for item in documents:
            if doc in item['number']:
                file = item
                break
        print((f"Владелец документа: {file['name']}"))
###--------------------------------------    
def shelf_search():
    file = input('Введите номер документа ')
    if doc_checking(file) == 0:
        print('Такого документа нет')
    else:
        for item,values in directories.items():
            if file in values:
                print(f'Документ расположен на полке {item}')
###--------------------------------------

def add_doc():
    doc = str(input('введите номер документа '))
    if doc_checking(doc) == 1:
        print('Такой документ существует')
    else:
        type_doc = input('введите тип документа ')
        name_doc = input('введите владельца документа ')
        shelf_doc = str(input('введите номер полки для хранения '))  
        if shelf_checking(shelf_doc) == 0:
            print('Такой полки не существует. Добавьте полку командой ads') 
        else:
            directories[shelf_doc].append(doc)
            documents.append({'type': type_doc, 'number': doc, 'name': name_doc})
            print('Документ добавлен')
        
    print(f'Текущий список документов : \n')
    list_print()
###--------------------------------------

def delete_shelf():
    
    shelf_del = str(input('Введите номер полки '))

    if shelf_checking(shelf_del) == 0:
        print(f"Такой полки не существует. Текущий перечень полок: {(','.join(list(directories.keys())))}")
    else:
        if len(directories[shelf_del]) != 0:
            print(f"На полке есть документ, удалите их перед удалением полки. Текущий перечень полок: {(','.join(list(directories.keys())))}")
        else:
            del(directories[shelf_del])
            print(f"Полка удалена. Текущий перечень полок: {(','.join(list(directories.keys())))}")
###--------------------------------------

def delete_doc():    
    doc = input('введите номер документа ')
    if doc_checking(doc) != 1:
        print('Документ не найден в базе. Текущий список документов:')
    else:
        for num, row in enumerate(documents):
            if row['number'] == doc:
                del(documents[num])
        for shelf in directories:
            if doc in directories[shelf]:
                doc_index = directories[shelf].index(doc)
                directories[shelf].pop(doc_index)
        print('Документ удален. Текущий список документов:')        
    list_print()
###--------------------------------------
def quit():
    print("Работа завершена")


commands = {
    'p': people,
    's': shelf_search,
    'l': list_print,
    'ads': add_shelf,
    'ds':delete_shelf,
    'd': delete_doc,
    'm': move_doc,
    'ad': add_doc,
    'q': quit
}


def start():
    command = ''
    while command != 'q':
        command = input("введите команду ")
        if not command in commands.keys():
            print("такой команды нет")
        else:
            commands[command]()
            print('')       

start()
  
```

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").
